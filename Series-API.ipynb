```python
final = pd.DataFrame()
for j in range(1,100):
    
    url = "https://api.themoviedb.org/3/tv/top_rated?language=en-US&page=1"

    headers = {
        "accept": "application/json",
        "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiI4OTVlZWQzZWYyZWI0NjVjYzk5YWNlYjZiYzc1ZWVkYyIsInN1YiI6IjY0N2RmNzQ1MGZiMzk4MDBkZTY4NGVlMCIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.O8GJUvGv1gjLEc9FNrrpgCovWOh_XPgHfLLvwVUD3LQ"
    }

    response = requests.get(url, headers=headers).json()
    popularity = []
    vote_count = []
    original_language = []
    name = []
    vote_average = []

    for i in response['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_language.append(i['original_language'])
        name.append(i['name'])
        vote_average.append(i['vote_average'])
    d = {'name':name, 'original_language':original_language,'popularity':popularity,
        'vote_count':vote_count,'vote_average':vote_average}
    df = pd.DataFrame(d)
    final = final.append(df, ignore_index=True)
```
